```python
fig, ax = plt.subplots()
ax.scatter(df['Gender'], df['Timestamp']) 
ax.set_xlabel("Gender")
ax.set_ylabel("Timestamp")
```

Analysis : The gender in all the frames is Male
We know that here there in only one person throughout the video who is a Male.
Prediction is correct

```python
fig, ax = plt.subplots()
ax.scatter(df['Eyeglasses'], df['Timestamp']) 
ax.set_xlabel("Eyeglasses")
ax.set_ylabel("Timestamp")
```

Analysis : The person is not wearing Eyeglasses in any of the frames.
We know that here there in only one person throughout the video who is not wearing glasses.
Prediction is correct

```python
fig, ax = plt.subplots()
ax.scatter(df['EyesOpen'], df['Timestamp']) 
ax.set_xlabel("EyesOpen")
ax.set_ylabel("Timestamp")
```

Analysis: The person's eye is open in many frames.
While there are quite a few frames with eyes closed too.

```python
fig, ax = plt.subplots()
ax.scatter(df['Smile'], df['Timestamp']) 
ax.set_xlabel("Smile")
ax.set_ylabel("Timestamp")
```

Analysis: The person is not smiling in most of the frames.
there are very few instaces where he is smiling.

```python
sns.relplot(x='Dominant emotion',y='Timestamp',data=df)
```

Analysis : In most of the frames the emotion exhibited by the person is CALM and the emotion least exhibited is that of CONFUSED. 

```python
fig, ax = plt.subplots()
ax.scatter(df['Pose'], df['Timestamp']) 
ax.set_xlabel("Pose")
ax.set_ylabel("Timestamp")
sns.lmplot(x='Pose',y='Timestamp',data=df)
```

Analysis : The predicted values are mostly between 15-25.

```python
sns.lmplot(x='AgeRange(Low)',y='Timestamp',data=df)
sns.lmplot(x='AgeRange(High)',y='Timestamp',data=df)
```

```python
sns.relplot(x='AgeRange(High)',y='AgeRange(Low)',hue='Timestamp',data=df)
```

Analyis : The predicted age range is mostly around 27-47.
